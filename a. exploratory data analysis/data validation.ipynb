```python
import pandas as pd

df = pd.read_csv("C:/Users/loydt/Documents/datasets/processed_df.csv")
```

We need to get the unique values of each column and use them for data validation. Values that don't conform to the expected patterns for each variable are considered invalid.

```python
# Unique values
unique_values = {col: df[col].unique() for col in df.columns}
print(unique_values)
```

```python
# Check numerical columns
num_col = [col for col in df.columns if pd.api.types.is_numeric_dtype(df[col])]
print(num_col)
```

```python
# Check min and max values for numerical columns
min_max_val = {col: {'min': df[col].min(), 'max': df[col].max()} for col in num_col}
for col, values in min_max_val.items():
    print(f"Column '{col}': min = {values['min']:.2f}, max = {values['max']:.2f}")
```

```python
# Categorical columns and their unique values
cat_col = [col for col in df.columns if pd.api.types.is_object_dtype(df[col])]
cat_col = {col: df[col].unique() for col in cat_col}
cat_col
```

```python
date_max = df['purchase_date'].max()
date_min = df['purchase_date'].min()

print(date_max)
print(date_min)
```
